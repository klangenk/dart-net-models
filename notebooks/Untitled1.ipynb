```python
def get_stats(path):
    paths = glob(f"{path}/*/")
    result = {}
    for p in paths:
        name = os.path.basename(os.path.dirname(p))
        result[name] = len(glob(f"{p}**/*.jpg", recursive=True) + glob(f"{p}**/*.jpeg", recursive=True))
    return result
```

```python
def cleanup(path):
    for key, value in get_stats(path).items():
        if value < 10: shutil.rmtree(f"{path}/{key}")
    
```

```python
def remove_null(path):
    # Read in the file
    with open(path, 'r') as file :
        filedata = file.read()

    # Replace the target string
    filedata = filedata.replace('\x00', '')

    # Write the file out again
    with open(path, 'w') as file:
        file.write(filedata)
```

```python
def read_csv(path):
    with open(path, newline='') as inFile:
        spamreader = csv.reader(inFile, delimiter=',', quotechar='"')
        result = []
        for row in spamreader:
            if row[0] != 'id':
                result.append(row)
    return result
```

```python
def check(path):
    max_id = None
    max_id_check = None
    if not os.path.isfile(f'{path}/darts.csv'): return None
    remove_null(f'{path}/darts.csv')
    with open(f'{path}/darts.csv', newline='') as inFile:
        reader = csv.reader(inFile, delimiter=',', quotechar='"')
        for row in reader:
            if row[0].isnumeric(): max_id = int(row[0])
    
    if not os.path.isfile(f'{path}/darts.csv'): return (max_id, max_id_check)
    remove_null(f'{path}/darts_check.csv')
    with open(f'{path}/darts_check.csv', newline='') as inFile:
        reader = csv.reader(inFile, delimiter=',', quotechar='"')
        for row in reader:
            if row[0].isnumeric(): max_id_check = int(row[0])
            
    return (max_id, max_id_check)
```

```python
def move_checked():
    source = '../data/real/1-captured/Webcam'
    paths = glob(f"{source}/*/")
    for p in paths:
        result = check(p)
        print(p, result)
        if result is not None and result[1] is not None and result[1] / result[0] > 0.9:
            #print(p)
            #print(result)
            shutil.move(p, p.replace('1-captured', '2-checked'))
    
```
